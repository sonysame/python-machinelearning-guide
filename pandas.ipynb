## DataFrame과 리스트, 딕셔너리, 넘파이 ndarray 상호 변환
### DataFrame을 넘파이 ndarray, 리스트, 딕셔너리로 변환
- DataFrame을 넘파이 ndarray로 변환하는 것은 DataFrame 객체의 values를 이용해 쉽게 할 수 있다. 
- 넘파이 ndarray를 list로 변환할 때는 tolist()를 사용
- 딕셔너리로의 변환은 to_dict()를 사용, 인자로 list를 입력하여 딕셔너리의 값을 리스트로 반환

```python
array3=df_dict.values
print('df_dict.values 타입:',type(array3),'df_dict.values shape:',array3.shape)
print(array3)

list3=df_dict.values.tolist()
print('df_dict.values.tolist() 타입:', type(list3))
print(list3)

dict3=df_dict.to_dict('list')
print('df_dict.to_dict() 타입:', type(dict3))
print(dict3)
```

### DataFrame의 칼럼 데이터 세트 생성과 수정<br/>
- 삭제는 drop() 메서드 이용
- axis 0은 row 방향, axis 1은 column 방향 축

```python
titanic_df['Age_0']=0
titanic_df['Age_by_10']=titanic_df['Age']*10
titanic_df['Age_by_10']=titanic_df['Age_by_10']+100
titanic_df['Family_No']=titanic_df['SibSp']+titanic_df['Parch']+1
titanic_df.head(3)
```

labels에 원하는 column명을 입력하고 axis=1을 입력하면 지정된 column을 드롭한다.<br/>
axis=0을 입력하면 row 축 방향으로 dropp을 수행하므로 특정 row를 drop한다. 

```python
titanic_drop_df=titanic_df.drop('Age_0', axis=1)
titanic_drop_df.head(3)
```

```python
titanic_df.head(3)
```

Age_0이 살아있는데, 이는 inplace 파라미터가 default값인 false로 설정되어있기 때문이다.<br/>
inplace값을 True로 설정하면 자신의 DataFrame의 데이터를 삭제한다.

```python
drop_result=titanic_df.drop(['Age_0', 'Age_by_10', 'Family_No'], axis=1, inplace=True)
titanic_df.head(3)
```

## apply lambda 식으로 데이터 가공

```python
titanic_df['Name_len']=titanic_df['Name'].apply(lambda x: len(x))
titanic_df
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else 'Adult')
titanic_df
```

```python
titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x:'Child' if x<=15 else ('Adult' if x<=60 else 'Elderly'))
titanic_df['Child_Adult'].value_counts()
```

```python
def get_category(age):
    cat=''
    if age<=5: 
        cat="Baby"
    elif age<=12:
        cat="Child"
    elif age<=18:
        cat="Teenager"
    elif age<=25:
        cat="Student"
    elif age<=35:
        cat="Young Adult"
    elif age<=60:
        cat="Adult"
    else:
        cat="Elderly"
    return cat

titanic_df['Child_Adult']=titanic_df['Age'].apply(lambda x: get_category(x))
titanic_df['Child_Adult'].value_counts()
```
